```python
def reverse_number(n):
    # Convert to string, reverse, and convert back to int
    return int(str(n)[::-1])

# Example usage
num = 12345
reversed_num = reverse_number(num)
print(reversed_num)  # Output: 54321
```

```python
reverse_number(67890)  # Output: 9876
```
